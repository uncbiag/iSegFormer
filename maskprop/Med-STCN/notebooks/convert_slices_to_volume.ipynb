```python
gt_root = '/playpen-raid2/qinliu/data/AbdomenCT-1K/Organ-12-Subset/Mask'
seg_root = '/playpen-raid2/qinliu/projects/iSegFormer/maskprop/Med-STCN/results/ABD1K'

result_bl = f'{seg_root}/stcn'
result_ft = f'{seg_root}/stcn_med_abdomen1k_Aug01_15.34.08_ft_s012_10k_no_cc'
result_cc = f'{seg_root}/stcn_med_abdomen1k_Aug01_22.03.33_ft_s012_10k_cc'

import os
import SimpleITK as sitk
import cv2 
import numpy as np
import matplotlib.pyplot as plt


volume_idxes = [f'Organ12_000{i}'  for i in range(1,10)] + [f'Organ12_00{i}'  for i in range(10,41)]

for volume_idx in volume_idxes:
    print(volume_idx)

    gt_path = f'{gt_root}/{volume_idx}.nii.gz'
    gt = sitk.ReadImage(gt_path)

    gt_npy = sitk.GetArrayFromImage(gt)
    seg_npy = np.zeros_like(gt_npy)

    results = [f'{seg_root}/stcn_med_abdomen1k_Aug01_15.34.08_ft_s012_10k_no_cc', f'{seg_root}/stcn_med_abdomen1k_Aug01_22.03.33_ft_s012_10k_cc']
    for result in results:
        for label_idx in range(1, 13):
                
            folder = f'{result}/label_{label_idx}/{volume_idx}'

            slices = []
            slice_names_all = os.listdir(folder)
            slice_names_all.sort()
            for slice_name in slice_names_all:
                slice = cv2.imread(f'{folder}/{slice_name}')

                slice_r, slice_g, slice_b = slice[:,:,0], slice[:,:,1], slice[:,:,2]
                slice_r[slice_r > 0] = label_idx 
                slice_r[slice_g > 0] = label_idx 
                slice_r[slice_b > 0] = label_idx 
                slice = slice_r

                slice = cv2.resize(slice, (512, 512), interpolation=cv2.INTER_NEAREST)

                slices.append(slice)
            slices = np.stack(slices, axis=0)

            seg_npy[slices>0] = label_idx

            # seg = sitk.GetImageFromArray(slices)
            # sitk.WriteImage(seg, f'{result}/{volume_idx}_label_{label_idx}.nii.gz')

        seg_all = sitk.GetImageFromArray(seg_npy)
        seg_all.CopyInformation(gt)        

        sitk.WriteImage(seg_all, f'{result}/{volume_idx}_label_all.nii.gz')



        
```
