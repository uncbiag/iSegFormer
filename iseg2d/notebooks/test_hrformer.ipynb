```python
import sys
import torch

sys.path.insert(0, '..')
from isegm.model.modeling.hrformer import HRT_B_OCR_V3

import os
import torch.distributed as dist

def setup(rank, world_size):
    os.environ['MASTER_ADDR'] = 'localhost'
    os.environ['MASTER_PORT'] = '12359'

    # initialize the process group
    dist.init_process_group("gloo", rank=rank, world_size=world_size)

def cleanup():
    dist.destroy_process_group()

setup(0, 1)
```

```python
pretrained_path = '/playpen-raid/qinliu/projects/ritm_interactive_segmentation/weights/pretrained/hrt_base.pth'
state_dict = torch.load(pretrained_path)
```

```python
from collections import OrderedDict

weights = state_dict['model']
new_ckpt = OrderedDict()
for k, v in weights.items():
    new_ckpt['backbone.' + k] = v
```

```python
network = HRT_B_OCR_V3(2, 6)

ori_proj_weight = new_ckpt['backbone.conv1.weight']
new_ckpt['backbone.conv1.weight'] = torch.cat([ori_proj_weight, ori_proj_weight], dim=1)


network.load_state_dict(new_ckpt, False)
```
